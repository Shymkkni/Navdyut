ADAVANCED ANALYSIS


 Performing clustering analysis to group states with similar crime profiles

 Creating a composite "crime index" that combines the different crime rates

 Building a simple predictive model to estimate one crime type based on the others

 Analyzing regional patterns by grouping states by geographic region

 Identifing outliers

```python
from sklearn.cluster import KMeans

X = data[['Murder', 'Assault', 'UrbanPop', 'Rape']]

kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)

data[['State', 'Cluster']]
```

```python
data['Crime_Index'] = data[['Murder', 'Assault', 'UrbanPop', 'Rape']].mean(axis=1)

data[['State', 'Crime_Index']]
```

```python
from sklearn.linear_model import LinearRegression

X = data[['Murder', 'Rape', 'UrbanPop']]
y = data['Assault']

model = LinearRegression()
model.fit(X, y)

print("Model R^2 Score:", model.score(X, y))
```

```python
northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'New Jersey', 'Pennsylvania']
south = ['Delaware', 'Maryland', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Alabama', 'Mississippi', 'Tennessee', 'Kentucky', 'Louisiana', 'Arkansas', 'Texas', 'Oklahoma']
midwest = ['Ohio', 'Michigan', 'Indiana', 'Wisconsin', 'Illinois', 'Minnesota', 'Iowa', 'Missouri', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas']
west = ['Montana', 'Idaho', 'Wyoming', 'Colorado', 'New Mexico', 'Arizona', 'Utah', 'Nevada', 'California', 'Oregon', 'Washington', 'Alaska', 'Hawaii']

def assign_region(state):
    if state in northeast:
        return 'Northeast'
    elif state in south:
        return 'South'
    elif state in midwest:
        return 'Midwest'
    else:
        return 'West'

data['Region'] = data['State'].apply(assign_region)

data.groupby('Region')['Crime_Index'].mean()
```

```python
Q1 = data['Murder'].quantile(0.25)
Q3 = data['Murder'].quantile(0.75)
IQR = Q3 - Q1

outliers_murder = data[(data['Murder'] > (Q3 + (1.5 * IQR)))]

outliers_murder[['State', 'Murder']]
```

```python
columns = ['Murder', 'Assault', 'UrbanPop', 'Rape']

outlier_states = {}

for col in columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    Q0 = Q1 - 1.5 * IQR
    Q4 = Q3 + 1.5 * IQR

    low_outliers = data[data[col] < Q0]['State'].values
    high_outliers = data[data[col] > Q4]['State'].values
    
    outlier_states[col] = {
        'Low Outliers': low_outliers,
        'High Outliers': high_outliers
    }

for crime, result in outlier_states.items():
    print(f"\n🔹 {crime} Outliers:")
    print(f"   ➡️ Low Outliers: {result['Low Outliers']}")
    print(f"   ➡️ High Outliers: {result['High Outliers']}")
```
